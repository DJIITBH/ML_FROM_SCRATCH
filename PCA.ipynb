#LIBRARY

```python
import numpy as np
import matplotlib.pyplot as plt
```

#DATA

```python
data = np.array([[2.,5], [6,4], [10,11], [14,14]])
data
```

#MAKE DATA MEAN CENTRIC

```python
mean1 = []
mean1.append(np.mean(data[:,0]))
mean1.append(np.mean(data[:,1]))
data[:,0] = data[:,0] - mean1[0]
data[:,1] = data[:,1] - mean1[1]
plt.scatter(data[:, 0], data[:, 1], color='blue', marker='o')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

#FIND CO VARIANCE MATRIX

```python
x1 = np.array(data[:,0])
y1 = np.array(data[:,1])
b = x1*y1
d = x1@x1.transpose()/len(b)
e = y1@y1.transpose()/len(b)
c =0
for i in range(len(b)):
  c = c + b[i]
c = c/len(b)
cov = np.array([[d,c],[c,e]])
cov
```

#FIND EIGEN VECTOR - PRINCIPAL COMPONENTS

```python
value, vec = np.linalg.eig(cov)
print(value)
print(vec)
m = vec[1,0]/vec[0,0]
pts = []
for d in data:
  pts.append(vec[0,0]*d[0] + vec[1,0]*d[1])

plt.scatter(pts[:],[0 for _ in range(len(pts[:4]))],color='blue', marker='o')
# plt.scatter(pts[4:],[0 for _ in range(len(pts[4:]))],color='red', marker='x', label='X2')
```

```python
x = np.linspace(-10, 10, 100)  # 100 points between -10 and 10

# Calculate y values using the line equation y = mx + b
y = m * x

# Plot the line
plt.plot(x, y, label=f'y = {m}x', color='blue')
plt.scatter(data[:, 0], data[:, 1], color='red', marker='o')
# Optionally, plot the origin for reference
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of the Line with Given Slope')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
