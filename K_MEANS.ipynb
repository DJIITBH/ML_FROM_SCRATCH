#LIBRARIES

```python
import numpy as np
import matplotlib.pyplot as plt
```

#DATA

```python
data = np.array([[1,4], [1,3], [0,4], [5,1], [6,2], [4,0]])
data
plt.scatter(data[:, 0], data[:, 1], color='blue', marker='o')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

#FINDING CLUSTERS

```python
def dis(x, y):
  return np.sqrt((x[0] - y[0])**2 + (x[1] - y[1])**2)

c1 = data[0]
c2 = data[5]
k1 = []
k2 = []
# k1.append(c)
# k2.append(c2)
mean = np.array([data[2],data[3]])
mean_new = np.array([c1,c2])
while not np.array_equal(mean, mean_new):
  for d in data:
    if dis(d, mean[0]) < dis(d, mean[1]):
      k1.append(d)
    else:
      k2.append(d)
  mean = [mean_new[0], mean_new[1]]
  mean_new = [np.mean(k1,axis = 0), np.mean(k2, axis =0)]
  # print(k1)
# print(mean)
# print(mean_new)
k1 = np.array(k1)
k2 = np.array(k2)
mean_new = np.array(mean_new)
print(k1)
print(k2)
```

#VISUALIZE

```python
plt.scatter(mean_new[:, 0], mean_new[:, 1], color='green', marker='*')
plt.scatter(k1[:, 0], k1[:, 1], color='blue', marker='o', label='Cluster 1')
plt.scatter(k2[:, 0], k2[:, 1], color='red', marker='x', label='Cluster 2')
```
